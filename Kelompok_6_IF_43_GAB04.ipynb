# **Proses Menampilkan Dataset**

```
# Mengimport library yang akan digunakan 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Untuk menampilkan data 
url = 'https://raw.githubusercontent.com/krobus00/Tubes-Statistika/master/BankChurners%20(sakshigoyal7).csv'
df = pd.read_csv(url)
df.head()
```

```
# Melihat ukuran (baris dan kolom) dataset yang akan digunakan 
df.shape
```

# **Analisis dan Visualisasi Data**

**1. Statistik Deskriptif**

Kami membuat deskripsi data secara statistik agar kami lebih mengenal data yang akan kami olah. Statistik deskriptif membantu kami menampilkan deskripsi data secara ringkas dari bagaimana bentuk penyebaran atau distribusi datanya hingga ukuran pemusatan data tersebut. 

1.   Ukuran pemusatan data : count, mean, median, minimum, maximum
2.   Ukuran penyebaran data : kuartil 1, kuartil 2, kuartil 3, standar deviasi



```
# Melihat apakah terdapat data yang tidak memiliki nilai
df.isna()
```

```
# Menampilkan statistik deskriptif untuk data numerik
round(df.describe(),2) 
```

```
# Menampilkan statistik deskriptif untuk data kategorik
df.describe(exclude=np.number)
```

**2. Histogram**

Data yang disajikan pada histogram ini adalah umur customer. Kami memodelkan data berupa histogram untuk melihat bagaimana distribusi usia customer pada bank ini. Dari histogram yang telah kami buat di bawah ini, dapat ditarik kesimpulan bahwa penyebaran umur customer bank paling banyak berada di daerah 40-50 tahun.  

```
# Menampilkan histogram yang menggambarkan penyebaran umur customer
plt.hist(df['Customer_Age'], bins=6)
plt.title('Penyebaran Umur Customer', fontweight = 'bold')
plt.xlabel('Umur Customer')
plt.ylabel('Jumlah Customer')
plt.show()
```

**3. Pie Chart**

Data yang disajikan pada pie chart ini adalah tingkat pendidikan dari customer bank. Kami memodelkan data berupa pie chart untuk melihat bagaimana persentase tingkat pendidikan customer. Dari pie chart yang telah kami buat di bawah ini, dapat dilihat bahwa tingkat pendidikan customer terbanyak merupakan graduate customer dengan perolehan sebesar 30.89%.

```
# Menampilkan grafik pie untuk education level
sns.set_palette('coolwarm')
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
label=['Graduate', 'High School','Unknown','Uneducated','College','Post-Graduate','Doctorate']
ax.pie(df['Education_Level'].value_counts(), labels=label, autopct='%1.2f%%')
plt.show()
```

**4. Count Plot**

Kami menggunakan fungsi count plot untuk menghitung frekuensi data berdasarkan kategorinya. Data yang disajikan pada count plot ini adalah kategori pendapatan customer bank. Berdasarkan count plot yang telah kami buat di bawah ini, dapat dilihat bahwa sebagian besar customer bank memiliki pendapatan tidak lebih dari $40K. 

```
# Menampilkan jumlah customer berdasarkan kategori kartu
sns.countplot(df['Income_Category'])
plt.title('Jumlah Customer Berdasarkan Kategori Pendapatan')
plt.xlabel('Kategori Pendapatan', labelpad=20)
plt.tick_params(axis='x',rotation=45)
plt.ylabel('Jumlah Customer')
plt.figure(figsize=(8,5))
plt.show()
```

Data yang disajikan pada count plot ini adalah kategori kartu berdasarkan gender customer. Dari count plot yang telah kami buat dibawah ini, dapat disimpulkan bahwa jumlah kategori kartu yang paling banyak digunakan adalah blue card dengan customer wanita lebih banyak menggunakannya daripada customer pria. 

```
# Menampilkan jumlah customer pria dan wanita berdasarkan kategori kartu
sns.countplot(df['Card_Category'], hue=df['Gender'])
plt.title('Jumlah Customer Pria dan Wanita')
plt.xlabel('Kategori Kartu')
plt.ylabel('Jumlah Customer')
plt.figure(figsize=(8,5))
plt.show()
```

**5. Box Plot**

Box Plot berikut menyajikan karakteristik dari data kategori kartu dan total transaction amount. Box plot dibawah menunjukkan bahwa total transaction amount keempat kategori kartu yaitu Blue, Gold, Silver, dan Platinum memiliki median yang berbeda-beda. Selain itu, terlihat bahwa total transaction amount dari kategori kartu Platinum yang menyebar (memiliki rentang yang lebih besar) dibandingkan dari ketiga kategori kartu lainnya. Adapun total transaction amount dari keempat kategori kartu tersebut tidak simetris (condong) dengan arah condong yang berbeda-beda, karena median tidak berada di tengah box dan salah satu dari whisker lebih panjang dari yang lainnya. Pada box plot total transaction amount dari kategori kartu Blue memiliki pencilan(outlier) terdapat total transaction amount yang bernilai jauh diatas kuartil 3, sedangkan pada total transaction amount dari kategori kartu Gold, Silver, dan Platinum tidak memiliki pencilan.

```
# Menampilkan box plot yang menggambarkan karakteristik dari data kategori kartu dan total transaction amount
sns.boxplot(x=df['Card_Category'], y=df['Total_Trans_Amt']);
plt.title('Box Plot dari Kategori Kartu dan Total Transaction Amount')
plt.xlabel('Kategori Kartu')
plt.ylabel('Total Transaction Amount')
plt.show()
```

**6. Heat Map**

Kami menggunakan heat map untuk menunjukkan bagaimana hubungan korelasi pada semua data yang tersedia. Berdasarkan koefisien korelasi dan heatmap yang telah dibuat, kami dapat menarik beberapa kesimpulan :

1.   Nilai koefisien Total_Trans_Amt dengan Total_Revolving_Bal adalah sebesar 0.1 yang berarti memiliki hubungan lemah positif.
2.   Nilai koefisien Credit_Limit dengan Avg_Open_To_Buy adalah sebesar 1.0 yang berarti memiliki hubungan kuat positif.
3.   Nilai koefisien Credit_Limit dengan Total_Relationship_Count adalah sebesar -1.0 yang berarti memiliki hubungan lemah negatif.
4.   Nilai koefisien Avg_Utilization_Ratio dengan Total_Revolving_Bal adalah sebesar 0.6 yang berarti memiliki hubungan kuat positif.



```
fig,ax=plt.subplots()
fig.set_size_inches(17,10)
cat_cols = [1,3,5,6,21,22] # kolom yang berisi data kategori atau data yang tidak digunakan
num_cols = [e for e in range(0,len(df.columns)) if e not in cat_cols] # kolom yang berisi data numerik
sns.heatmap(data=df.iloc[:,num_cols].corr(), ax=ax, annot=True, fmt='.1f')
plt.show()
```

**7. Korelasi**

unnamed.png

Nilai korelasi antara Total Transaction Amount dengan Total Transaction Count adalah sebesar 0.81. Artinya, antara Total Transaction Amount dengan Total Transaction Count memiliki hubungan yang kuat karena nilai korelasi termasuk ke dalam golongan positif kuat.

```
# Menampilkan nilai korelasi antara Total Transaction Amount dengan Total Transaction Count
print("Nilai Korelasi = ", round(np.corrcoef(df["Total_Trans_Amt"],df["Total_Trans_Ct"])[1][0],2)) # Nilai korelasi akan dibulatkan 2 angka di belakang koma
```

Nilai korelasi antara Credit_Limit dengan Avg_Open_To_Buy adalah sebesar 1.0. Artinya, antara Credit_Limit dengan Avg_Open_To_Buy memiliki hubungan yang kuat karena nilai korelasi termasuk ke dalam golongan positif kuat.



```
# Menampilkan nilai korelasi antara Credit_Limit dengan Avg_Open_To_Buy
print("Nilai Korelasi = ", round(np.corrcoef(df["Credit_Limit"],df["Avg_Open_To_Buy"])[1][0],2)) # Nilai korelasi akan dibulatkan 2 angka di belakang koma
```

**8. Scatter Plot**

Scatter plot berikut menyajikan hubungan antara Total_Trans_Amt dengan Total_Trans_Ct. Berdasarkan koefisien korelasi dan bentuk scatter plot yang dihasilkan dibawah ini, dapat diketahui bahwa antara Total_Trans_Amt dengan Total_Trans_Ct memiliki hubungan kuat positif. Artinya, semakin tinggi jumlah transaksinya maka semakin tinggi pula total nominal transaksinya.

```
# Menampilkan scatter plot hubungan antara total transaction amount dengan total transaction count
sns.scatterplot(x=df['Total_Trans_Amt'], y=df['Total_Trans_Ct'])
plt.xlabel('Total Transaction Amount') 
plt.ylabel('Total Transaction Count')
plt.title('Hubungan Antara Total Transaction Amount dengan Total Transaction Count')
plt.show()
```

Scatter plot berikut menyajikan hubungan antara credit_limit dengan avg_open_to_buy. Berdasarkan koefisien korelasi dan bentuk scatter plot yang dihasilkan dibawah ini, dapat diketahui bahwa antara credit_limit dengan avg_open_to_buy memiliki hubungan kuat positif. Artinya, semakin tinggi limitasi kreditnya maka semakin tinggi pula rata-rata open to buy nya. 

```
# Menampilkan scatter plot hubungan antara Credit_Limit dengan Avg_Open_To_Buy
sns.scatterplot(x=df['Credit_Limit'], y=df['Avg_Open_To_Buy'])
plt.xlabel('Credit_Limit')
plt.ylabel('Avg_Open_To_Buy')
plt.title('Hubungan Antara Credit_Limit dengan Avg_Open_To_Buy')
plt.show()
```
